```python
import numpy as np
from skimage.io import imread
import matplotlib.pyplot as plt
from skimage.color import rgb2gray
```

```python
img=imread('https://images.unsplash.com/photo-1459411552884-841db9b3cc2a?ixlib=rb-1.2.1&ixid=MnwxMjA3fDB8MHxzZWFyY2h8MTJ8fHJhbmRvbXxlbnwwfHwwfHw%3D&auto=format&fit=crop&w=500&q=60')
img=rgb2gray(img)
plt.imshow(np.around(img,decimals=3),cmap='gray')
```

```python
flt=img.flatten()
len(flt)
```

```python
flt
```

```python
flt1=np.around(np.array(flt),decimals=3)
flt1=list(set(flt1))
flt1=list(flt1)
flt1.sort()
flt1
```

```python
flt=np.around(flt,decimals=3)
```

```python
len(flt1)
```

```python
type(flt1)
```

```python
hist=np.histogram(flt,bins=flt1)
```

```python
N=np.sum (hist[0])
```

```python
N
```

```python
plt.hist(flt,bins=flt1)
plt.show()
```

```python
P=hist[0]/N
P
```

```python
len(P)
```

```python
np.sum(P)
```

```python
variances=[]
for i in range(1,len(P)):
  w=np.sum(P[:i])
  u=0
  for j in range(1,i+1):
    u+=(j*P[j])
  ut=0
  for k in range(1,len(P)):
    ut+=(k*P[k])
  sig=pow(((ut*w)-u),2)/(w*(1-w))
  variances.append(sig)
```

```python
maxthres=variances.index( max(variances))
```

```python
maxthres
```

```python
backimg=img>flt1[maxthres]
plt.imshow(backimg,cmap='gray')
```
